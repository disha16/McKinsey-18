```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
data = pd.read_csv(r"C:\Users\Rawal Rawal\Desktop\Fremont_Bridge_Hourly_Bicycle_Counts_by_Month_October_2012_to_present.csv", index_col='Date', parse_dates=True)
data.columns = ['East','West']
data['Total']=data.eval('East+West')
data =data.resample('M').mean()
data.rolling(30, center=True).mean().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count');
plt.xlabel('Dates')
```
