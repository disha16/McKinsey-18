```python
import pandas as pd
import statsmodels.api as sm
import scipy
import matplotlib.pyplot as pyplot
from scipy import stats
import numpy as np
from pandas.core.nanops import nanmean as pd_nanmean
from statsmodels.regression.linear_model import OLS
from statsmodels.tsa.seasonal import seasonal_decompose
df=pd.read_csv("C:\Users\Rawal Rawal\Downloads\salesdata.csv")
df.index=pd.to_datetime(df['Date'])
series = df['Sales']
df.fillna(0)
result = seasonal_decompose(series, model='additive',extrapolate_trend=6)
result.resid = np.array(result.resid)
result.resid = map(lambda x:x+3000, result.resid)
result.seasonal = np.array(result.seasonal)
result.trend = np.array(result.trend)
trendeffects = result.trend[-24:]
seasonalityeffects = result.seasonal[-24:]
test =  df[-24:]
train = df[0:-24]
power = scipy.stats.boxcox_normmax(result.resid, method='mle')
result.resid = map(lambda x:x-3000, result.resid)
result4 = (result.resid**power - 1)/power
result4 = np.array(result4)
result4 = result4[:-24]
result4 = pd.DataFrame(result4, index=train.index,columns = ['val'])
forecast = []
Values = []
MAPE = []
Shapirovals = []
for p in [1,2,3,4,5]:
      for q in [0,1,2,3,4,5]:
            try:
                mod = sm.tsa.statespace.SARIMAX(result4['val'],order=(p,0,q))
                res = mod.fit(disp=False)
                forecasted = [res.forecast(steps=24).values]
                testvals = np.array([test['Sales']])
                forecasted = np.array(forecasted)
                forecasted = ((forecasted*0.95)+1)**(1/0.95)
                forecasted = forecasted+trendeffects+seasonalityeffects
                mapev = np.(testvals-forecasted)*100/testvals
                error = (testvals-forecasted)
                W,r = scipy.stats.shapiro(error)
                mape = np.mean(mapev)
                MAPE.append(mape)
                Shapirovals.append(r)
                Values.append([p,q])
                forecast.append(forecasted)
            except:
                continue
                
indexapp = np.argmax(Shapirovals)
forecasted = forecast[indexapp]
print  (MAPE[indexapp])
print (Shapirovals[indexapp])
print  (Values[indexapp])
print (forecasted)




```
