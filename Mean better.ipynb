```python
import numpy
import pandas as pd
import matplotlib as pyplot
path = raw_input("Your CSV file: ")
n = raw_input("Moving Average of:  ")
n = int(n)
df = pd.read_csv(path)
df.set_index('date')
df['value2'] =df['value'].copy()
df['rolling_mean']=df['value2'].rolling(window=n).mean(center=True)
rn = raw_input("Second Order for Moving Average:  ")
rn=int(rn)
df['rolling_mean']=df['value2'].rolling(window=rn).mean(center=True)
df['rolling_mean'] = df['rolling_mean'].shift(1-rn)
df.plot(x='date', y=['value', 'rolling_mean'], figsize=(10,5), grid=True)
```
